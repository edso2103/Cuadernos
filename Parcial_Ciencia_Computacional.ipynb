#<FONT FACE="times new roman" SIZE=4> <b>1) Juan y María acordaron encontrarse en la Universidad y estudiar para el parcial de ciencia computacional intermedia entre las 12 m y la 1 pm. Los dos llegan en cualquier momento dentro de ese intervalo de tiempo. Suponiendo que los tiempos de llegada de cada uno de ellos son independientes, calcular la probabilidad de que Juan y María se encuentren, sabiendo que María solo espera cinco minutos y Juan solo 20</b></FONT>





##<FONT FACE="times new roman" SIZE=5>**Solución analítica**
x: Juan 
y: Maria

1.  Definimos el espacio muestral como:  $\Omega=\{{ (x,y) ∈ R; 0 ≤ x ≤ 60,  0 ≤ y ≤ 60 }\}$

2. Definimos el evento $T_M$ como la diferencia entre las hora de llegada que sea menor o igual a  5 en el caso de María:  
$\{T_M= { (x,y) ∈ \Omega\ ; 0 ≤ x-y ≤ 5\}}$  
<br>
Y definimos el evento $T_J$ como la diferencia entre las hora de llegada que sea menor o igual a 20 en el caso de Juan:  
$\{T_J= { (x,y) ∈ \Omega\ ; 0 ≤ y-x ≤ 20}\}$

* Probabilidad de que se encuentren Maria y Juan

$$P(Encontrarse)=\frac{m(A)}{m(Ω )}$$  

$\{m(Ω)\}$ = Área total del espacio muestral (cuadrado)  

$\{m(Ω)\}$= Lado*Lado  

$\{m(Ω)\}$= $60*60$ = $3600$  


$\{m(A)\}$ = Área de la figura en la que Juan y María se encuentran 

$\{m(A)\}$= Área total -(Área triangulo 1 + Área triangulo 2) 'Probabilidad Geométrica'

$\{m(A)\}$= $3600-((\frac{55*55}{2})+(\frac{40*40}{2}))$ 

$\{m(A)\}$= $1287,5$  
  
  Por lor tanto: $$P(Encontrarse)=\frac{1287,5}{3600}= \frac{103}{288} \approx0.36 $$

Juan y Maria se encontrarían el 36% de las veces

</FONT>


##<FONT FACE="times new roman" SIZE=5>**Solución simulada**

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import random
```

```python
from numpy import random
lanzamientos = 10000
Juan = list()
Maria = list()
for i in range(lanzamientos):
    Juan.append(random.uniform(0,60))
    Maria.append(random.uniform(0,60))
print(Juan[:15])
print()
print(Maria[:15])
```

```python
estimado = [resultados.count('(Se encontraron)')/lanzamientos,
            resultados.count('(No se encontraron)')/lanzamientos]

print(f'Frecuencia relativa: {estimado}')
print(f'Intento totales: {lanzamientos}')   
```

#<FONT FACE="times new roman" SIZE=4><b>2) Se tienen dos urnas A y B. La urna A contiene 3 bolas rojas y 4 negras y la urna B contiene 5 bolas rojas y 7 negras. Se selecciona al azar una bola de cada urna. ¿A que es igual la probabilidad de que las dos bolas seleccionadas sean del mismo color?</b></FONT>

##<FONT FACE="times new roman" SIZE=4>**Solución analítica**
Definimos el espacio muestral como $\Omega=\{rr,rn,nn,nr\}$

$$P(rr)= \frac{3}{7} * \frac{5}{12}=\frac{5}{28}\approx 0.18$$  
$$P(nn)= \frac{4}{7} * \frac{7}{12}=\frac{1}{3}\approx 0.33$$  

$$P(rr~∪~nn)= \frac{5}{28} + \frac{1}{3}=\frac{43}{84}\approx 0.51$$

La probabilidad de que dos bolas seleccionadas sean del mismo color es de 51%

</FONT>



##<FONT FACE="times new roman" SIZE=4>**Solución simulada**


```python
lanzamientos=10000

urnaA = ['roja','roja','roja', 'negra','negra','negra','negra']
urnaB = ['roja','roja','roja', 'roja', 'roja', 'negra','negra','negra','negra','negra','negra','negra']
probabilidad_tiempoIgual=list() 
probabilidad_tiempoDiferente=list() 
resultados=list()

urnaA_resultado=list()
urnaB_resultado=list()
probabilidadIgual=0
probabilidadDiferente=0


for i in range(lanzamientos):
     urnaA_resultado.append(random.choice(urnaA))
     urnaB_resultado.append(random.choice(urnaB))

     if ((urnaA_resultado[i]=='roja' and urnaB_resultado[i]=='roja') or (urnaA_resultado[i]=='negra' and urnaB_resultado[i]=='negra')):
        probabilidadIgual+=1
        resultados.append('Iguales')
     else:
        probabilidadDiferente+=1
        resultados.append('Diferentes')

     probabilidad_tiempoIgual.append(resultados.count('Iguales')/(i+1))
     probabilidad_tiempoDiferente.append(resultados.count('Diferentes')/(i+1))
      

print("Urna A:", urnaA_resultado[:5])
print("Urna B:", urnaB_resultado[:5])
print("Probabilidad de obtener bolas de mismo color:", probabilidadIgual/lanzamientos)
print("Probabilidad de obtener bolas de diferente color:", probabilidadDiferente/lanzamientos)
print("Probabilidad de obtener bolas de mismo color por lanzamiento", probabilidad_tiempoIgual[:5])
print("Probabilidad de obtener bolas de diferente color por lanzamiento", probabilidad_tiempoDiferente[:5])
```

```python
estimado = [(resultados.count('Iguales'))/lanzamientos, 
            (resultados.count('Diferentes')/lanzamientos)]
print(f'Frecuencia relativa: {estimado}')
print(f'Intentos totales: {lanzamientos}')
```

#<FONT FACE="times new roman" SIZE=4>**3) Se ordenan 5 hombres y 5 mujeres de acuerdo con sus calificaciones en un examen. Suponga que no hay dos calificaciones iguales y que los 10! órdenes posibles son igualmente probables. A que es igual la probabilidad de que la posición más alta alcanzada por un hombre sea la cuarta.**</FONT>


##<FONT FACE="times new roman" SIZE=4>**Solución analítica**

Definimos el espacio muestral como $\Omega=\{Hombre,~Mujer\}$  
<br>
**Evento A**= probabilidad de que la posición más alta alcanzada por un hombre sea la cuarta.

$$P(A)=\frac{Casos~favorables}{(Casos~totales)}$$
**Casos favorables:**


tabla.PNG
<br>
$$P(A)=\frac{5*4*3*5*6!}{10!}=\frac{216000}{3628800}=\frac{5}{84} = 0.0595 \approx 0.06$$

La probabilidad de que la posición más alta alcanzada por un hombre sea la cuarta es de 6%
</FONT>

##<FONT FACE="times new roman" SIZE=4>**Solución simulada**


```python
from numpy import random
lanzamientos=20000
probabilidad_tiempoSI=list() 
probabilidad_tiempoNO=list() 
resultados=list()

probabilidadSI=0
probabilidadNO=0

Estudiantes=["M","M","M","M","M","H","H","H","H","H"];          

for i in range (lanzamientos):
  random.shuffle(Estudiantes)
  if ( (Estudiantes[0] is 'M') and (Estudiantes[1] is 'M') and (Estudiantes[2] is 'M') and (Estudiantes[3] is 'H')):
    probabilidadSI+=1
    resultados.append('SI')
  else:
    probabilidadNO+=1
    resultados.append('NO')
  
  probabilidad_tiempoSI.append(resultados.count('SI')/(i+1))
  probabilidad_tiempoNO.append(resultados.count('NO')/(i+1))

print("Probabilidad de que se encuentre en la posicion 4:", probabilidadSI/lanzamientos)
print("Probabilidad de que no se encuentre en la posicion 4:", probabilidadNO/lanzamientos)
print("Probabilidad de que este el hombre con la calificacion mas alta en la posicion 4 por lanzamiento", probabilidad_tiempoSI[10:20])
print("Probabilidad de que no este el hombre con la calificacion mas alta en la posicion 4 por lanzamiento", probabilidad_tiempoNO[10:20])
```

#<FONT FACE="times new roman" SIZE=4>**4) En la tienda de la Sergio Arboleda venden dos tipos de frutas: Manzanas y Peras. Se ha comprobado que la tercera parte de los estudiantes eligen Manzanas y el resto prefiere comer Peras. Además, la probabilidad de encontrar una Manzana que esté en mal estado es del 0.05 y la de encontrar una Pera en mal estado es de 0.09. ¿Cuál es la probabilidad de que un estudiante compre una fruta y esté en mal estado? ¿Cuál es la probabilidad de que una fruta sea una Pera sabiendo que está en mal estado?**</FONT>

##<FONT FACE="times new roman" SIZE=4>**Solución analítica**
1.  Se define el espacio muestral como:  Ω={peras, manzanas}

$P(M_{anzana})$ = Elije manzana

$P(P_{era})$ = Elije pera

$P(M_e|M_{anzana})$ = En mal estado dado que elije manzana 

$P(M_e|P_{era})$ = En mal estado dado que elije pera

$P(M_{anzana})$ = $\frac{1}{3}$

$P(P_{era})$ = $\frac{2}{3}$

$P(M_e|M_{anzana})$ = $0.05$

$P(M_e|P_{era})$ = $0.09$

**1.** Probabilidad de que un estudiante compre una fruta y esté en mal estado

$$P(M_e) = P(M_e∩M_{anzana}) + P(M_e∩P_{era})$$
<br>

$$P(M_e) = P(M_{anzana})*P(M_e|M_{anzana})+P(P_{era})*P(M_e|P_{era})$$
<br>
$$P(M_e) = \frac{1}{3}*0.05+\frac{2}{3}*0.09 = \frac{23}{300} \approx 0.077$$
<br>
La probabilidad de que un estudiante compre una fruta y este en mal estado es de 7.7%
<br>
<br>
**2.** Probabilidad de que una fruta sea una Pera sabiendo que está en mal estado

$$P(P_{era}|M_e) = \frac{P(M_e∩P_{era})}{P(M_e)}$$
<br>
$$P(P_{era}|M_e) = \frac{\frac{2}{3}*0.09}{\frac{1}{3}*0.05+\frac{2}{3}*0.09}=\frac{18}{23} \approx 0.7826$$

La probabilidad de que una fruta sea pera sabiendo que esta en mal estado es de 78.26%

</FONT>


##<FONT FACE="times new roman" SIZE=4>**Solución simulada**


```python
from numpy import random

lanzamientos = 10000

#Generar numeros aleatorios entre 0 y 1
evento1=random.rand(lanzamientos) #pera o manzana
evento2=random.rand(lanzamientos) #mal estado de pera o manzana
print("Evento 1: ",evento1[:5])
print("Evento 2:",evento2[:5])
```

```python
resultados=list()
probTFrutaMal = []
probTPeraMal = []

for i in range(lanzamientos):
  # Manzana(tercera parte de los estudiantes) o pera(el resto)
  #Manzana
  if (evento1[i]>=0 and evento1[i]<=1/3):
    #Probabilidad del mal estado dado que escogió manzana (ME|M) = 0.05
    if (evento2[i]>=0 and evento2[i]<=0.05):
      resultados.append("Manzana")
      resultados.append("mal estado")

  #Pera
  elif (evento1[i]>1/3 and evento1[i]<=1):

     #Probabilidad del mal estado dado que escogió pera (ME|P) = 0.09
    if (evento2[i]>=0 and evento2[i]<=0.09):
      resultados.append("Pera")
      resultados.append("mal estado")

  #una fruta sea una pera sabiendo que está en mal estado?
  probTPeraMal.append(resultados.count("Pera")
                      /(resultados.count("mal estado")+1))
  #compre una fruta y esté en mal estado?
  probTFrutaMal.append(resultados.count("mal estado")/(i+1))
  
print("P(encontrar una Pera en mal estado):\n",probTPeraMal[-20:])
print("\nP(encontrar una fruta en mal estado):\n",probTFrutaMal[-20:])
```

```python
estimado = [resultados.count("mal estado")/lanzamientos,
            resultados.count("Pera")/resultados.count("mal estado")]

print(f'\nFrecuencia relativa: {estimado}')
print(f'Intentos totales: {lanzamientos}')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(30, 10))

plt.subplot(1,2,1)
plt.bar(["Pera dado que la fruta esta en mal estado","Fruta en mal estado"], estimado, color='turquoise')
plt.title("Comprar una fruta en la Universidad Sergio Arboleda")
plt.xlabel("Eventos")
plt.ylabel("Probabilidad")
plt.subplot(1,2,2)
plt.plot(range(1, lanzamientos+1), probTFrutaMal, label="Fruta en mal estado")
plt.plot(range(1, lanzamientos+1), [23/300]*lanzamientos, color='red')
plt.plot(range(1, lanzamientos+1),probTPeraMal, label="Pera dado que la fruta esta en mal estado")
plt.plot(range(1, lanzamientos+1), [18/23]*lanzamientos, color='red')
plt.title('Cambio de la probabilidad estimada en el tiempo')
plt.ylabel('Probabilidad')
plt.xlabel('Tiempo')
plt.legend()

plt.show()
```

#<FONT FACE="times new roman" SIZE=4>**6. Los coeficientes A, B y C de la ecuación cuadrática  $Ax^2+Bx+C=0 $ son determinados lanzando un dado corriente tres veces consecutivas.** 




##<FONT FACE="times new roman" SIZE=4>**Solución analítica**  
<br>
$$P(A)=\frac{Casos~favorables}{(Casos~totales)}$$
<br>

**1.** ¿A qué es igual la probabilidad de que las dos raíces de la ecuación
sean reales? 
<br>

**Casos totales:** Debido a que se lanza 3 veces un dado (con 6 opciones) para obtener el valor de A, B Y C, se tiene que los casos totales son:  
$$Casos~totales= 6*6*6=216$$  
<br>
**Casos favorables:**
Corresponden al caso en el que la solución de las raíces es real, lo que implica que de la ecuación cuadrática,  
$$X= \frac{-b-\sqrt{b^2-4*a*c}}{2*a}$$  
en la sección,  
$$b^2-(4*a*c)≥0$$  
debería ser mayor o igual a 0, ya que es lo que se encuentra dentro de la raíz y determina si es compleja o no, y al realizar la resta, *b* debe ser mayor al producto de 4,a y c,  

$$b^2≥(4*a*c)$$
<br>
Al sumar los valores que corresponden a esta condición se obtiene que:  
(Tenga en cuenta que se cambian los valores de a y c)
<br>
b=1 -> número de raices reales=0  
b=2 -> número de raices reales=1  
b=3 -> número de raices reales=3  
b=4 -> número de raices reales=8  
b=5 -> número de raices reales=14  
b=6 -> número de raices reales=17  
<br>
Obteniendo así, **43 raices reales**.  
$$P(raices~reales)=\frac{43}{216}\approx 0.199$$

**2.** ¿A qué es igual la probabilidad de que las dos raíces sean complejas? Encuentre la solución
analítica y la solución simulada

Al obtener el numero de raices reales,se obtiene que:  

$$raices~complejas=casos~totales-raices ~reales$$  
$$raices~complejas=216-43=173$$  
<br>  
$$P(raices~complejas)=\frac{173}{216}\approx 0.80$$</FONT>


##<FONT FACE="times new roman" SIZE=4>**Solución simulada**


```python
import numpy as np 

lanzamientos = 300
a = [1,2,3,4,5,6]
b = [1,2,3,4,5,6]
c = [1,2,3,4,5,6]
```

```python
resultados = []
total = 0
probTRaizR = []
probTRaizC = []
random.shuffle(a)
random.shuffle(b)
random.shuffle(c)
for l in range(lanzamientos):
  for i in b:
    for j in a:
      for k in c:
        total += 1
        if i**2 >= (4*j*k):
          resultados.append("Real")
        else:
          resultados.append("Compleja")
        probTRaizR.append(resultados.count("Real")/total)
        probTRaizC.append(resultados.count("Compleja")/total)

print("Probabilidad de encontrar una raiz real:\n",probTRaizR[:15])
print("Probabilidad de encontrar una raiz compleja:\n",probTRaizC[:15])
  
```

```python
estimado = [resultados.count("Real")/total,
            resultados.count("Compleja")/total]

print(f'\nFrecuencia relativa: {estimado}')
print(f'Intentos totales: {total}')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(30, 10))

plt.subplot(1,2,1)
plt.bar(["Raiz Real","Raiz Compleja"], estimado, color='turquoise')
plt.title("Lanzar un dado 3 veces (a,b,c)")
plt.xlabel("Eventos")
plt.ylabel("Probabilidad")

plt.subplot(1,2,2)
plt.xlim(0,2000)
plt.plot(range(1, total+1), probTRaizR, label="Raiz Real")
plt.plot(range(1, total+1), [43/216]*total, color='red')
plt.plot(range(1, total+1), probTRaizC, label="Raiz Compleja")
plt.plot(range(1, total+1), [173/216]*total, color='red')
plt.title('Cambio de la probabilidad estimada en el tiempo')
plt.ylabel('Probabilidad')
plt.xlabel('Tiempo')
plt.legend()

plt.show()
```
