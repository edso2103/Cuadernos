# <FONT FACE="times new roman" SIZE=4> **¿Qué es inteligencia de enjambre?**  



<FONT FACE="times new roman" SIZE=4> Estos algoritmos se centran en analizar los comportamientos colectivos y la interación de los individuos entre si. En la actualidad se conocen varios sistemas de enjambre como colonia de hormigas, bandadas de aves, banco de peces e incluso luciérnagas, pero para la implementación del algoritmo se elegirá la optimización de enjambre de partículas.



<FONT FACE="times new roman" SIZE=4> La optimización del enjambre de partículas (PSO),es un algotitmo de optimización global  consiste en que todos los componentes ubiquen un objetivo común, para que luego se vayan explorando distintas posiciones y vayan avanzando gradualmente hacia ciertas posiciones.  
Para este algoritmo, se tienen en cuenta soluciones candidatas llamadas "partículas" que se mueven por todo un espacio de búsqueda teniendo en cuenta la velocidad y la posición de la partícula, con el fin de que lleguen rápidamente hacia las mejores soluciones 



<FONT FACE="times new roman" SIZE=4> Estos algoritmos se formalizan con:
*  f(x): Una función de costo a minimizar, que recibe como argumento la solución candidata
*  $S$: La nube de partículas 
*  $x_i$: La posición de la partícula
*  $v_i$: La velocidad de la partícula
*  $p_i$: La mejor posición de la partícula
*  $g$: La mejor posición global


<FONT FACE="times new roman" SIZE=4> El procedimiento consiste en:
1.  Inicializar las posiciones de las partículas por medio de un vector con valores aleatorios, que se encuentren uniformente distribuido, es decir con las mismas probabilidades de aparecer. Y también se debe inicializar la mejor posición conocida, $(p_i)$.  
2. Posterior a ello, se realizan comparaciones, donde si $f(xi)$ es menor que $f(p_i)$ entonces se actualiza el valor de $p_i$, y si $f(p_i)$ es menor que $f(g)$ entonces se actualiza el valor de $g$, hasta llegar a la mejor solución.  
3. Es importante tener en cuenta que a medida que se realiza este procedimiento se deben ir actualizando las velocidades y las posiciones de las partículas.  
4. Además, en estos algoritmos, se hace uso de parámetos w, $φ_p$, $φ_g$ los cuales regulan la dependencia que tienen las particulas y con ello la eficacia del algoritmo PSO.  
5. Para finalizar, se recomienda usar mayor número de partículas y de iteraciones para encontrar mejores soluciones.

# <FONT FACE="times new roman" SIZE=4> **Implementación**  


```python
#Se importan las bibliotecas 
#random: valores aleatorios
#numpy: arreglos
import random
import numpy as np 

#Se define la función que ajusta la posición de las particulas 
#Se busca el punto mínimo de la función, es decir, (0,0) 
#por lo que entre menores sean x y y mejor será el ajuste.
#La función corresponde a f(x,y)=x²+y²+1
def funcion_ajuste(pos):
    return pos[0]**2 + pos[1]**2 + 1

#Se inicializan los parámetros que definen la velocidad
W = 0.5 #Parámetro de inercia que afecta el movimiento
c1 = 0.5 #Coeficiente de aceleración para el p_i
c2 = 0.9 #Coeficiente de aceleración para el g
target = 1

#Se pide el número de iteraciones, el número de partículas y el error
num_iteraciones = int(input("Digite el número de iteraciones: "))
error = float(input("Digite el margen de error soportado: "))
num_particulas = int(input("Digite el número de partículas: "))

#Se generan el vector de posiciones aleatorias para las partículas
#El vector cuenta con posiciones (x,y) de valores entre -50 y 50
#Se encuentra uniformemente distrubuido, 
vector_x_particula = np.array([np.array([(-1) ** (bool(random.getrandbits(1)))
                     * random.random()*50, (-1)**(bool(random.getrandbits(1))) 
                     * random.random()*50]) for _ in range(num_particulas)])

#Se inicializa la mejor posición de la partícula
mejor_posicion_p = vector_x_particula

#Se inicializa la variables min_posicion p_i y g, en el mayor valor, inf
#Para las partículas, un vector para organizar las posiciones
min_posicion_p = np.array([float('inf') for _ in range(num_particulas)])
min_posicion_g = float('inf')

#se inicializa la mejor solución global (x,y)
mejor_posicion_g = np.array([float('inf'), float('inf')])

#Se define el vector de las velocidades de las partículas
vector_velocidades = ([np.array([0, 0]) for _ in range(num_particulas)])

#Se define la variable iteración para el ciclo
iteracion = 0

#Se recorre por el número de iteraciones digitadas
while iteracion < num_iteraciones:
    #Se recorren todas las partículas
    for i in range(num_particulas):
        #Se evalúa la función de ajuste, según la posición de cada partícula
        actual = funcion_ajuste(vector_x_particula[i])
        print(actual, ' ', vector_x_particula[i])
        #Si la mejor posición encontrada es mayor que la actual
        #actualice el valor
        if(min_posicion_p[i] > actual):
            min_posicion_p[i] = actual 
            mejor_posicion_p[i] = vector_x_particula[i]

        #Se repite el proceso para los globales 
        if(min_posicion_g > actual):
            min_posicion_g = actual
            mejor_posicion_g = vector_x_particula[i]

    #Se verifica que el error de la solución este dentro del error digitado 
    if(abs(min_posicion_g - target) < error):
        break

    #Se recorren nuevamente las partículas
    for i in range(num_particulas):
      #Se actualiza la nueva posición y velocidad de las partículas
      #Se tienen en cuenta los coeficientes y las mejores posiciones encontradas
      #Se aplica la ecuación de velocidad: 
      #w*v(k)+c1*rand(p_i-x_i)+ c2*rand(p_g-x_i)
        ec_1=(W*vector_velocidades[i]) + (c1*random.random())
        ec_2=(mejor_posicion_p[i] - vector_x_particula[i]) +(c2*random.random())
        ec_3=(mejor_posicion_g-vector_x_particula[i])
        velocidad =  ec_1*ec_2*ec_3
        posicion = velocidad + vector_x_particula[i]
        vector_x_particula[i] = posicion
    #Se aumenta el contador
    iteracion = iteracion + 1
    
print("\nLa mejor posición global es: ", mejor_posicion_g)
```
