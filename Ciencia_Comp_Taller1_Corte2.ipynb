##<FONT FACE="times new roman" SIZE=4>**Solución**

**A.** Función de densidad (derivar la función de distribución)

$$f_X(x)=\frac{dF_X(x)}{dx}$$  
$$\frac{dF_{X}(x)}{dx}=2-2x$$
$$f_X(x)=2-2x$$

**B.** Calcular: 

**P(X ≥ 3/2):** Dado que cuando x (3/2) es mayor a 1 la probabilidad es 1.

$$1-1=0$$

**P(−2 ≤ X ≤ 3/4):** Dado que cuando x es menor a 0 la probabilidad es de 0, entonces se evalua la función de distribución cuando (0 ≤ X ≤ 3/4).  

$$(2(\frac{3}{4})-(\frac{3}{4})^2)-(2(0)-(0)^2)$$
$$=0.9375$$

##<FONT FACE="times new roman" SIZE=4>**Solución**  
<br>

b: blanca  (-3000)  
n: negra  (+5000)  
a: azul

Definimos el espacio muestral como $\Omega=\{bb,nn,aa,ba,ab,an,na,nb,bn\}$

| X    | Evento | P(X=x) 
|         ----- | ---             | -----    
|-6000 |bb |  $$\frac{8}{16}*\frac{7}{15}=0.233$$ | 
|-3000 |ba,ab | $$(\frac{8}{16}*\frac{3}{15})*2=0.2$$  | 
| 0 |aa |  $$\frac{3}{16}*\frac{2}{15}=0.025$$ |

\begin{equation*}

\end{equation*}


**B.** Función de distribución  =  

\begin{equation*}
F_X(x)=p(X≤x)=\left\{ \begin{array} \\0.233,&  -6000≤ x<-3000\\ 0.433,& -3000≤ x< 0\\0.458,& 0≤ x< 2000\\0.78,& 2000≤ x< 5000\\0.91,&5000≤ x < 10000\\1,& x ≥10000 \end{array}\right.

#<FONT FACE="times new roman" SIZE=4>**7. Se tienen tres urnas A, B y C. La urna A contiene 5 bolas rojas y 4 blancas, la urna B contiene 8 bolas rojas y 5 blancas y la urna C contiene 2 bolas rojas y 6 blancas. Se extrae al azar una bola de cada urna. Sea X el número de bolas blancas extraídas. Calcular las funciones de densidad y de distribución para la variable aleatoria X.**

##<FONT FACE="times new roman" SIZE=4>**Solución**  
<br>

r=roja  
b=blanca  


Definimos el espacio muestral como $\Omega=\{bbb,brb,bbr,rbb,rrr,rbr,rrb,brr\}$  
X: Bolas blancas



| X    | Evento | P(X=x) 
|         ----- | ---             | -----    
|0|rrr |  $$\frac{5}{9}*\frac{8}{13}*\frac{2}{8}=0.0854$$ | 
|1 |rbr,rrb,brr | $$(\frac{5}{9}*\frac{5}{13}*\frac{2}{8})+(\frac{5}{9}*\frac{8}{13}*\frac{6}{8})+(\frac{4}{9}*\frac{8}{13}*\frac{2}{8})=0.378$$  | 
| 2 |brb,bbr,rbb|  $$(\frac{4}{9}*\frac{8}{13}*\frac{6}{8})+(\frac{4}{9}*\frac{5}{13}*\frac{2}{8})+(\frac{5}{9}*\frac{5}{13}*\frac{6}{8})=0.408$$  |
|3 |bbb|  $$\frac{4}{9}*\frac{5}{13}*\frac{6}{8}=0.128$$ |  
  
<BR>

**A.** Función de densidad  = 
\begin{equation*}
f_X(x)=F'(x)=\left\{ \begin{array} \\0.0854,&  P(X=0) \\ 0.378,& P(X=1)\\0.408,& P(X=2) \\0.128,& P(X=3) \end{array}\right.
\end{equation*}

**B.** Función de distribución  = 
\begin{equation*}
F_X(x)=p(X≤x)=\left\{ \begin{array} \\0.0854,&  0≤ x< 1 \\ 0.4634,& 1≤ x< 2\\0.87714,& 2≤ x< 3 \\1,& x ≥3 \end{array}\right.
\end{equation*}



</FONT>
