```
from IPython.display import display
import ipywidgets as widgets
import time
import random
import json

class Tablero:
    def __init__(self, tamano_celda=(50, 50), n_celdas=(),objetos=[]):
        self.out = widgets.HTML()
        display(self.out)
        self.tamano_celda = tamano_celda
        self.n_celdas = n_celdas
        self.objetos = objetos

    def generate_garbage(self):
        for x_1 in range(self.n_celdas[0]):
            for y_1 in range(self.n_celdas[1]):
                n = random.randint(0,10)
                if n>5:
                    garbage = Agente(x_1, y_1, 0, emoticon=f"üçÇ{x_1,y_1}", tamano_emoticon=15)
                    objetos.insert(0,garbage)
                    
    def interaccion(self, agent):
        agent.set_agentes_env([n for n in self.objetos if not n == agent])
        
        
    def dibujar(self):
        tablero = "<table border='1' >{}</table>"
        filas = ""

        for j in range(self.n_celdas[0]):
          s = ""
          for i in range(self.n_celdas[1]):
            contenido =""
            for o in objetos:
              if o.x == j and o.y == i:
                contenido = \
                "<div style='transform: rotate({angulo}deg);font-size:{tamano_emoticon}px;'>{emoticon}</div>".\
                format(angulo = o.angulo, tamano_emoticon = o.tamano_emoticon, emoticon = o.emoticon)     
            s += "<td style='height:{alto}px;width:{ancho}px'>{contenido}</td>".\
              format(alto = self.tamano_celda[0], ancho = self.tamano_celda[1], 
                    contenido = contenido)
          filas += "<tr>{}</tr>".format(s)
        tablero = tablero.format(filas)
        self.out.value = tablero
```

```
class Agente(object):
    def __init__(self, x, y, angulo, emoticon, tamano_emoticon=30):
        self.x = x
        self.y = y
        self.angulo = angulo
        self.emoticon = emoticon
        self.tamano_emoticon = tamano_emoticon
    
```

```
class Vacuum(Agente):
    def __init__(self, x=0, y=0, angulo=0, emoticon="ü§ñ", tamano_emoticon=30, energia=5):
        Agente.__init__(self,x,y,angulo,emoticon,tamano_emoticon)
        self.energia = energia
        self.agentes_cerca=[]
        self.agentes_env = []
        self.agentes_aspiraos= []
        self.energia_dict = {"energia por moverse":0,"energia por deteccion":0,"energia por aspirar":0}
        
    def set_agentes_env(self, obj):
        self.agentes_env = obj

    def move_wih_garbage(self,limits):
        if not self.agentes_cerca == []:
            aux = (self.agentes_cerca[0].x-self.x,self.agentes_cerca[0].y-self.y)
            if aux == (1,0):
                self.angulo = 0
            elif aux == (0,1):
                self.angulo = 270
            elif aux == (-1,0):
                self.angulo=180
            elif aux == (0,-1):
                self.angulo = 90
            else:
                #print("Move with garbage")
                if self.angulo == 0:
                    self.angulo = 270
                else:
                    self.angulo -=90
           
    def move(self,env):
        limit = env.n_celdas
        self.move_wih_garbage(limit)
        
        if self.angulo == 0 and self.energia > 0 and limit[0] > self.x+1 and self.x>=0:
            self.x+=1
            self.energia -= 1
            self.energia_dict["energia por moverse"] +=1
        elif self.angulo == 90 and self.energia > 0 and limit[1] > self.y and self.y-1>=0:
            self.y-=1
            self.energia -= 1
            self.energia_dict["energia por moverse"] +=1
        elif self.angulo == 180 and self.energia > 0 and limit[0] > self.x and self.x-1>=0:
            self.x -= 1
            self.energia -= 1
            self.energia_dict["energia por moverse"] +=1
        elif self.angulo == 270 and limit[1] > self.y+1 and self.energia > 0 and self.y>=0:
            self.y += 1
            self.energia -= 1
            self.energia_dict["energia por moverse"] +=1
        else:
            #print("Move with not if")
            #print(self.energia)
            if self.angulo == 0:
                self.angulo = 270
            else:
                self.angulo -=90
            
    def aspirar(self):
         for i in self.agentes_cerca:
            aux = (i.x-self.x,i.y-self.y)

            if aux == (0,0) and self.energia > 0:
                self.energia -= 1
                self.energia_dict["energia por aspirar"] +=1
                self.agentes_aspiraos.append(i)
                self.agentes_cerca.pop(self.agentes_cerca.index(i))
                
                return i
    
    def revisar(self):
        pos_valid = [(0,1),(1,0),(0,0)]
        self.agentes_cerca=[]
        for i in self.agentes_env:
            aux = (abs(i.x-self.x),abs(i.y-self.y))
            if aux in pos_valid and self.energia > 0:
                temp = (0,1) [aux == (0,0)] 
                self.energia -= temp
                self.energia_dict["energia por deteccion"] +=temp
                self.agentes_cerca.append(i)
```

```
repeticiones = {}

for rep in range(50): 
    agente = Vacuum(x=0,y=0, energia = 100)
    objetos = [agente]
    escenario = Tablero(n_celdas = (random.randint(1,10),random.randint(1,10)), objetos = objetos)
    #escenario = Tablero(n_celdas = (9,9), objetos = objetos)
    escenario.generate_garbage()
    energy_i = agente.energia
    helper = len(escenario.objetos)-1
    energy_u=agente.energia
    while(energy_u!=0):
        escenario.dibujar()
        escenario.interaccion(agente)
        agente.revisar()
        x = agente.aspirar()
        if x :
            escenario.objetos.pop(escenario.objetos.index(x)) 
        agente.move(escenario)
        #print(escenario.objetos)
        if len(escenario.objetos)== 1:
            break
        energy_u=agente.energia
        #print("Energia: ", energy_u)

    df = {"dimesiones":str(escenario.n_celdas),
          "energia utilizada":((energy_i-agente.energia)/energy_i),
          "info energia": agente.energia_dict,
          "basura faltante":(helper-len(agente.agentes_aspiraos))/helper if helper>0 else 0}
    repeticiones[rep] = df
```

```
energy_list = []
garbage_list=[]
for i,t in repeticiones.items():
    energy_list.append(t["energia utilizada"])
    garbage_list.append(t["basura faltante"])
promedio_energia = sum(energy_list)/len(energy_list)
promedio_basura = sum(garbage_list)/len(garbage_list)

print(promedio_energia)
print(1-promedio_basura)
```
